```
from wordcloud import WordCloud
import pandas as pd   
import matplotlib.pyplot as plt  
```

```
#Import Dataset
from google.colab import files
files.upload()
```

```
df=pd.read_csv("covid+death.csv")
print(df.shape)
df.head()
```

```
df.tail()
```

```
#Show All the reasons
df.groupby(["Condition"]).count()
```

```
df.groupby(["Condition Group"]).count()
```

```
# Create WordCloud
selist=df["Condition"].tolist()
```

```
selist_as_String = ' '.join(selist)
```

```
plt.figure(figsize=(14,14))
plt.imshow(WordCloud().generate(selist_as_String))
```

```
# Create WordCloud
selist1=df["Condition Group"].tolist()
```

```
selist_as_String1 = ' '.join(selist1)
```

```
plt.figure(figsize=(10,10))
plt.imshow(WordCloud().generate(selist_as_String1))
```
